Data Loading and Initial Exploration

```python
# Importing necessary libraries for data analysis and visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
df = pd.read_csv("/content/Books_Data_Clean.csv")# Adjust filename as needed
```

```python
df.head() # Displays the first five rows
```

```python
df.describe() # Displays basic dataset information to understand structure
```

Data Cleaning and Preprocessing


```python
df = df[df["Publishing Year"] > 1900]
# Filtering out records with unrealistic publishing years (pre-1900)
# This removes potential data entry errors or irrelevant historical records
```

```python
df.describe()
# Verifying the impact of filtering on statistical distribution
```

```python
df.isna().sum()
# Checking for missing values across all columns
# Identifying null values is crucial for data quality assessment
```

```python
df.dropna(subset = "Book Name" )
# Removing records with missing book names as they're essential for analysis
```

```python
# Checking for duplicate records to ensure data integrity
# Returns the count of duplicate rows
df.duplicated().sum()
 # Result: 0 duplicates found
```

```python
df.nunique()
# Counts unique values in each column to understand data diversity
```

Revenue Feature Engineering


```python
# Creating Revenue column
df["Revenue"] = df["sale price"] * df["units sold"]

df[["sale price", "units sold", "Revenue"]].head()
```

 Revenue Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Revenue"], bins=30, kde=True)
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

Revenue by Genre

```python
genre_revenue = df.groupby("genre")["Revenue"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
genre_revenue.plot(kind="bar")
plt.title("Total Revenue by Genre")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()

genre_revenue.head()
```

 Correlation Analysis


```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

Top 10 vs Bottom 10 Revenue Books


```python
# Sort by Revenue in descending order and get top 10 highest revenue books
top10 = df.sort_values("Revenue", ascending=False).head(10)

# Sort by Revenue in ascending order and get bottom 10 lowest revenue books
bottom10 = df.sort_values("Revenue", ascending=True).head(10)

# Display top 10 revenue-generating books with key metrics
print("=" * 80)
print("TOP 10 HIGHEST REVENUE BOOKS")
print("=" * 80)
print(top10[["Book Name", "genre", "Book_ratings_count", "Revenue"]])

print("\n" + "=" * 80)
print("BOTTOM 10 LOWEST REVENUE BOOKS")
print("=" * 80)
print(bottom10[["Book Name", "genre", "Book_ratings_count", "Revenue"]])
```

Exploratory Data Analysis (EDA)

1. Publishing Year Distribution


```python
# Visualizing the distribution of publishing years
# This helps identify trends in book publication over time
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of the Publishing Year")
plt.show()
```

2. Genre Analysis

```python
# Analyzing genre distribution using a bar chart
# Identifies the most common book genres in the dataset
df["genre"].value_counts().plot(kind = "bar")
plt.xlabel("Genre")
plt.ylabel("Frequency")
plt.title("Distribution of the Genre")
plt.show()
```

3. Author Performance Analysis

```python
# Calculating average book ratings per author
# Identifies top-performing authors based on reader ratings
df.groupby("Author")
["Book_average_rating"].mean().sort_values(ascending = False)
```

4. Genre vs Ratings Count

```python
# Comparing book ratings count across different genres using box plots
# Box plots reveal median, quartiles, and outliers for each genre
sns.boxplot(
    x = "genre",
    y = "Book_ratings_count",
    data = df,
    palette=["red","blue","green","purple"]
    )
plt.xlabel("Genre")
plt.ylabel("Book Ratings Count")
plt.title("Box Plot of Book Ratings Count for Each Genre")
plt.show()
```

5. Price vs Sales Relationship

```python
# Investigating the relationship between sale price and units sold
# Scatter plot helps identify pricing strategies and demand patterns
plt.scatter(df["sale price"], df["units sold"])
plt.xlabel("Sale Price")
plt.ylabel("Units Sold")
plt.title("Scatter Plot of Sale Price vs Units Sold")
plt.show()
```

6. Language Distribution

```python
# Counting books by language and sort in ascending order
language_counts = df["language_code"].value_counts().sort_values(ascending=True)

# Visualizing language distribution with a pie chart
# Shows the proportion of books available in each language
plt.pie(language_counts, labels=language_counts.index, startangle=90, autopct="%1.1f%%")
plt.axis("equal")
plt.title("Language Distribution of Languages")
plt.show()
```

9. Rating vs Popularity Correlation

```python
# Examining correlation between book ratings and popularity
# Higher ratings may correlate with more reader engagement
plt.scatter(df["Book_average_rating"], df["Book_ratings_count"])
plt.xlabel("Book Average Rating")
plt.ylabel("Book Ratings Count")
plt.title("Scatter Plot of Book Average Rating vs Book Ratings Count")
plt.show()
```

10. Top Authors by Sales Revenue

```python
# Calculating total gross sales per author
total_gross_sales_by_author = df.groupby("Author")["gross sales"].sum()
total_gross_sales_by_author
```

```python
# Visualizing top 20 authors by total gross sales
# Bar chart highlights the most commercially successful authors
total_gross_sales_by_author.sort_values(ascending = False).head(20).plot(kind = "bar")
plt.xlabel("Author")
plt.ylabel("Total Gross Sales")
plt.title("Total Gross Sales by Author")
plt.show()
```

11. Author Rating vs Sales Volume

```python
# Comparing units sold across different author rating categories
# Box plots reveal how author reputation affects sales performance
sns.boxplot(
    x="Author_Rating",
    y="units sold",
    data=df,
    palette=["red", "blue", "green", "purple"]
)
plt.xlabel("Author Rating")
plt.ylabel("Units Sold")
plt.title("Box Plot of Units Sold for Each Author Rating")
plt.show()
```

12. Sales Trends Over Time

```python
# Tracking total units sold by publishing year
# Line chart with markers shows sales trends and market evolution
df.groupby("Publishing Year")["units sold"].sum().plot(kind="line", marker="o")
plt.xlabel("Publishing Year")
plt.ylabel("Total Units Sold")
plt.title("Total Units Sold Over Time")
plt.show()
```
